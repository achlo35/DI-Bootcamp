```python
import random

def throw_dice():
    return random.randint(1, 6)

    
```

```python
def throw_until_double():
    count = 0
    while True:
        die1 = throw_dice()
        die2 = throw_dice()
        count += 1
        # print(f'Throw {count}: ({die1} , {die2})')
        if die1 == die2:
            break
    return count
    
total_throws = throw_until_double()
print(f'Total throws until doubles: {total_throws}')

```

```python
def main():
    results = []

    for i in range(100):
        throws = throw_until_double()
        results.append(throws)
        total_throws = sum(results)
        average_throws = total_throws / len(results)
        print(f"Total throws: {total_throws}")
        print(f"Average throws to reach doubles: {average_throws:.2f}")

main()
```
