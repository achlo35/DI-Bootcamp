```python
# Exercise 1 : Temperature

# Base class
class Temperature:
    def __init__(self, value):
        self.value = value

    def to_celsius(self):
        raise NotImplementedError

    def to_fahrenheit(self):
        raise NotImplementedError

    def to_kelvin(self):
        raise NotImplementedError

# Celsius subclass
class Celsius(Temperature):
    def to_celsius(self):
        return self.value

    def to_fahrenheit(self):
        return self.value * 9/5 + 32

    def to_kelvin(self):
        return self.value + 273.15

# Fahrenheit subclass
class Fahrenheit(Temperature):
    def to_celsius(self):
        return (self.value - 32) * 5/9

    def to_fahrenheit(self):
        return self.value

    def to_kelvin(self):
        return (self.value - 32) * 5/9 + 273.15

# Kelvin subclass
class Kelvin(Temperature):
    def to_celsius(self):
        return self.value - 273.15

    def to_fahrenheit(self):
        return (self.value - 273.15) * 9/5 + 32

    def to_kelvin(self):
        return self.value

# Example usage:
c = Celsius(25)
print("25°C to F:", c.to_fahrenheit())
print("25°C to K:", c.to_kelvin())

f = Fahrenheit(77)
print("77°F to C:", f.to_celsius())
print("77°F to K:", f.to_kelvin())

k = Kelvin(300)
print("300K to C:", k.to_celsius())
print("300K to F:", k.to_fahrenheit())


```

```python
# Exercise 2: In the Quantum Realm

import random

class QuantumParticle:
    def __init__(self, x, y, p):
        self.x = x
        self.y = y
        self.p = p
        self.entangled_with = None
        self.spin_value = None  # Will be set on measurement

    def __repr__(self):
        return f"QuantumParticle(x={self.x}, y={self.y}, p={self.p:.4f})"

    def disturbance(self):
        self.x = random.randint(1, 10000)
        self.y = random.randint(1, 10000)
        self.p = random.uniform(0, 1)
        print("Quantum Interferences!!")

    def position(self):
        self.disturbance()
        return self.x

    def momentum(self):
        self.disturbance()
        return self.p

    def entangle(self, other):
        if not isinstance(other, QuantumParticle):
            print("Can only entangle with another QuantumParticle!")
            return
        self.entangled_with = other
        other.entangled_with = self
        self.spin_value = None
        other.spin_value = None
        print("Spooky Action at a Distance !!")

    def spin(self):
        # If not entangled, just return a random spin
        if self.entangled_with is None:
            self.spin_value = random.choice([0.5, -0.5])
            return self.spin_value
        # If entangled and spin not yet measured, set both
        if self.spin_value is None and self.entangled_with.spin_value is None:
            self.spin_value = random.choice([0.5, -0.5])
            self.entangled_with.spin_value = -self.spin_value
            return self.spin_value
        # If this particle's spin is set, return it
        if self.spin_value is not None:
            return self.spin_value
        # If the other particle's spin is set, set this one to opposite
        if self.entangled_with.spin_value is not None:
            self.spin_value = -self.entangled_with.spin_value
            return self.spin_value

# Example usage:
p1 = QuantumParticle(1, 2, 0.7)
p2 = QuantumParticle(3, 4, 0.2)

p1.entangle(p2)  
print("p1 spin:", p1.spin())
print("p2 spin:", p2.spin())  
print("p1 spin again:", p1.spin())  




```
